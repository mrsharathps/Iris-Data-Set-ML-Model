# Modeling - Simple model

Decide some cutoff points for classification.

Based on EDA, if the petal length is less than 2.5cm, then clearly it reflects setosa species

```python
def single_feature_prediction(petal_length):
    """Predict the Iris species """
    if petal_length<2.5:
        return 0
    elif petal_length<4.8:
        return 1
    else:
        return 2
```

```python
def plot_incorrect_prediction(df_predictions, x_axis_feature, y_axis_feature):
    fig, axs = plt.subplots(2,2,figsize=(10,10))
    #Grid of 2, 2 with values till 10
    axs=axs.flatten()
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, hue="prediction_label", data=df_predictions, ax=axs[0]) 
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, hue="target_names", data=df_predictions, ax=axs[1])
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, hue="correct_prediction", data=df_predictions, ax=axs[2])
    axs[3].set_visible(False)
    plt.show()
```

```python
plot_incorrect_prediction(df_predictions, "petal length (cm)", "petal width (cm)")
```

# Model Tuning

Model tuning is trying to teremine the parameter of our model that maximizes the model performace. By all means we should be avoiding our model from memorizing any of the data.

 C : float, default=1.0
    Inverse of regularization strength; must be a positive float.
    Like in support vector machines, smaller values specify stronger
    regularization.

Giving lesser values will be more effective.

In our case increasing the value of C is not regularizing the model.

```python
for reg_param in (0.5,0.9,1,1.3,1.4,2,3,4,5,6,7):
    model=LogisticRegression(max_iter=200, C=reg_param)
    accuracies=cross_val_score(model,X_train,y_train,cv=5,scoring="accuracy")
    print(f"Accuracy : {np.mean(accuracies)*100: .2f}%")
```
